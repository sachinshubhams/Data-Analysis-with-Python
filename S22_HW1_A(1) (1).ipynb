## Add new columns with labels "mean" and "date" to the DataFrame df.

- ### Column "mean": mean of the 12 windspeeds recorded at 12 locations each day

- ### Column "date": date in the format 19YY-MM-DD. this column is derived from the first three columns of the dataframe df. you will need the Pandas function to_datetime()

## Generate two subplots. 

- ### Subplot 1 contains mean weekly windspeed at Loc 2 smoothened using rolling average window of 3 weeks.

- ### Subplot 2 contains median monthly windspeed at Loc 2 smoothened using rolling average window of 4 months.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12, 6)) 
#Mean weekly windspeed at Loc2
ax1 = fig.add_subplot(121)
df.Loc2.resample('W').mean().rolling(3).mean().plot(title="Mean weekly windspeed at Loc2")
#Median monthly windspeed at Loc2
ax2 = fig.add_subplot(122)
df.Loc2.resample('M').median().rolling(4).mean().plot(title="Median monthly windspeed at Loc2")
#show
plt.show()
```

## Generate a box plot representing windspeeds in Location 1 for each month

```python
df['Month']=df.index.month

fig, ax = plt.subplots(figsize=(15,10))
#set palette as bright
palette = sns.color_palette("bright")
#set palette and style
sns.set(style="whitegrid", palette=palette)
#plot boxplot
sns.boxplot(x="Month", y="Loc1", data=df)
```
